```python
#Ejemplo de los animales

class Animal():
    nombre = ""
    def comer(self):
        print("El animal comer")
    def dormir(self):
        print("El animal duerme")
#para linficar la herencia se usan los () despues del nombre de la clase
# class NOMBRE_CLASE(CLASE_PADRE)
class Mamifero(Animal):
    npezones = 0

    def amamantar(self):
        print("El animal amamanta a su cria")

    def metodo(self):
      print("Metodo en mamifero")

mamifero = Mamifero()
#metodos heredados
mamifero.comer()
mamifero.dormir()
#metodo propio
mamifero.amamantar()
```

```python
#Clase animal domestico

class AnimalDomestico():
  def pasear(self):
    print("El animal pasea")
  def metodo(self):
      print("Metodo en Animal Domestico")

#Podemos poner la clase perro que herede de Mamifero
#para indicar qeu heredamos de mas de un padre lo separamos por comas
#si tenemos varios metodos iguales en los padres si se tiene en cunta el orden
# Va revisando en el orden que los ponemos en la lista
class Perro(AnimalDomestico, Mamifero):
  def ladrar(self):
    print("El perro ladra")

perro = Perro()
#Herdedaso de su Abuelo que se Anima
perro.comer()
perro.dormir()
#heredado de su padre Mamifero
perro.amamantar()
#heredado de su otro parde animal domestico
perro.pasear()
#Metodo propio
perro.ladrar()
#Cual de los dos se ejecuta o se ejecuan ambos?
perro.metodo()
```

```python
#Estamos creand una clase llamada A
class A():
  def llover(self):
    print("Chove en galicia")
#aqui cremaos una clase llamada B que hereda de la clase A
class B(A):
  def llover(self):
    #podemos invodar al metodo original el del padre
    super().llover()
    #Segudo metodo con la clase directamente
    A.llover(self)
    print("No llueve en galicia")

b = B()
b.llover()
```

```python
#Simulacion de objetos
def calculadora(operacion):
  #funcion para sumar
  def suma(a,b):
    return a + b
  #funcion para restar
  def resta(a,b):
    return a - b
  if(operacion == "suma"):
    #devuelve la funcion suma
    return suma
  else:
    #devuelve la funcion resta
    return resta
#este primer ejemplo devuelve la funcion suma
operacion = calculadora("suma")
print(operacion)
print(operacion(5,5))
#en este caso la funcion resta
operacion_2 = calculadora("resta")
print(operacion_2)
print(operacion_2(5,5))
```

```python
#Recursividad funciones que se llaman a si mismas
#la recursividad es como hacer un bucle, por ejemplo un while
def contar(numero):
  #dentro de la propia funcion la vuelvo a invocar (recursividad)
  print(numero)
  contar(numero - 1)

contar(10)
```

```python
#al igual que en un bucle debemos poner una condicion de parada, es decir que deje
#de invocarse para que no sea infinito, esto se llama caso base
def contar(numero):
  #dentro de la propia funcion la vuelvo a invocar (recursividad)
  print(numero)
  #caso base aqui cuando avle 0 se deja de llamar a si misma
  if numero ==0:
    return 0
  else:
    contar(numero - 1)

contar(10)
```

```python
#Factorial de un numero
# 5! = 5 * 4 * 3 * 2 * 1
#TODO debemos crear la funcion factorial iterativa con un bucle (for)
#teneis 5 minutos

def factorial(numero):
  resultado = 1
  for i in range(numero,0,-1):
    #resultado =1
    #i = 5
    #primera vuelta resultado = 1 * 5 = 5
    #resultado = 5
    #i = 4
    #segund vuelta resultado = 5 * 4  = 20
    #resultado = 20
    #i = 3
    #tercera vuelta resultado = 20 * 3 = 60
    resultado = resultado * i

  return resultado

print(factorial(5)) #120
```

```python
#en recursivo

def factorial(numero):

  if numero == 1:
    return 1
  else:
    return numero * factorial(numero - 1)

print(factorial(5))
```

```python
class NumerosPositivos():
    __valor = 0
    def setValor(self, nuevo_valor):
        if nuevo_valor > 0:
            self.__valor = nuevo_valor
        else:
            print("NO ES POSITIVO")
    def getValor(self):
      return self.__valor

num = NumerosPositivos()
num.setValor(-1000)
print(num.getValor())
```

```python
class NumerosPositivos():
    __valor = 0
    #un decorados es una funcion que sirve para modificar otra funcion
    @property
    #esto se comporta como un getter
    def valor(self):
      return self.__valor
    @valor.setter
    def valor(self, nuevo_valor):
        if nuevo_valor > 0:
            self.__valor = nuevo_valor
        else:
            print("NO ES POSITIVO")


num = NumerosPositivos()
num.valor =-1000
print(num.valor)
```
